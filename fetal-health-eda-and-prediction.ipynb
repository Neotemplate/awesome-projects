<h1>About this dataset</h1>
<blockquote>
<h2>Context</h2>
<p>Reduction of child mortality is reflected in several of the United Nations' Sustainable Development Goals and is a key indicator of human progress.<br />The UN expects that by 2030, countries end preventable deaths of newborns and children under 5 years of age, with all countries aiming to reduce underâ€‘5 mortality to at least as low as 25 per 1,000 live births.</p>
<p>Parallel to notion of child mortality is of course maternal mortality, which accounts for&nbsp;<strong>295 000 deaths</strong>&nbsp;during and following pregnancy and childbirth (as of 2017). The vast majority of these deaths&nbsp;<strong>(94%)</strong>&nbsp;occurred in low-resource settings, and most&nbsp;<strong>could have been prevented</strong>.</p>
<p>In light of what was mentioned above,&nbsp;<strong>Cardiotocograms (CTGs)</strong>&nbsp;are a simple and cost accessible option to assess fetal health, allowing healthcare professionals to take action in order to prevent child and maternal mortality. The equipment itself works by sending ultrasound pulses and reading its response, thus shedding light on fetal heart rate (FHR), fetal movements, uterine contractions and more.</p>
<h2>Data</h2>
<p>This dataset contains&nbsp;<strong>2126</strong>&nbsp;records of features extracted from Cardiotocogram exams, which were then classified by three expert obstetritians into&nbsp;<strong>3 classes</strong>:</p>
<ul>
<li>Normal</li>
<li>Suspect</li>
<li>Pathological</li>
</ul>

```python
#Here, we are going to check with features hold mutual information with our target: fetal health. 

from sklearn.feature_selection import mutual_info_classif
def make_mi_scores(X, y):
    mi_scores = mutual_info_classif(X, y)
    mi_scores = pd.Series(mi_scores, name="MI Scores", index=X.columns)
    mi_scores = mi_scores.sort_values(ascending=False)
    mi_scores.plot.bar()
    return mi_scores

make_mi_scores(df, df.fetal_health)
```

```python
#data_to_scale = [i for i in df.columns if i not in ['fetal_health']]

#Prepare data for modeling. To avoid data leakage, first we split the data and then apply scaling to it.

X = df.copy()
y = X.pop('fetal_health')

#Make train and test split

from sklearn.model_selection import cross_val_score, train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=23)
```

<h2 style="text-align: center;"><span style="color: #003366; background-color: #ffffff;">SUPERVISED LEARNING WITH LGBM CLASSIFIER<img src="https://html-online.com/editor/tiny4_9_11/plugins/emoticons/img/smiley-cool.gif" alt="cool" /></span></h2>

<h3>Advantages of Light GBM</h3>
<ul>
<li><strong>Faster training speed and higher efficiency:</strong>&nbsp;Light GBM uses a histogram-based algorithm i.e it buckets continuous feature values into discrete bins which fasten the training procedure.</li>
<li><strong>Lower memory usage:</strong>&nbsp;Replaces continuous values to discrete bins which results in lower memory usage.</li>
<li><strong>Better accuracy than any other boosting algorithm:</strong>&nbsp;It produces much more complex trees by following leaf wise split approach rather than a level-wise approach which is the main factor in achieving higher accuracy.</li>
<li><strong>Compatibility with Large Datasets:</strong>&nbsp;It is capable of performing equally well with large datasets with a significant reduction in training time as compared to XGBoost.</li>
</ul>
<h3>Disadvantages of Light GBM</h3>
<ul>
<li><strong>Overfitting:</strong>&nbsp;Light GBM split the tree leaf-wise which can lead to overfitting as it produces much complex trees.</li>
<li><strong>Compatibility with Datasets:</strong>&nbsp;Light GBM is sensitive to overfitting and thus can easily overfit small data.</li>
</ul>

```python
from lightgbm import LGBMClassifier
from sklearn.metrics import classification_report, ConfusionMatrixDisplay, confusion_matrix

def model_predict(model):
    m = model(boosting_type='gbdt', num_leaves=31, max_depth=- 1, 
              learning_rate=0.5, n_estimators=100, subsample_for_bin=200000,
              objective=None, class_weight=None, min_split_gain=0.0, min_child_weight=0.001,
              min_child_samples=20, subsample=1.0, subsample_freq=0, colsample_bytree=1.0, 
              reg_alpha=0.0, reg_lambda=0.0, random_state=None, n_jobs=None, importance_type='split')
    
    m.fit(X_train, y_train)
    pred = m.predict(X_test)
    print(f'--Classification Report--\n\n{classification_report(y_test, pred)}')
    cm = confusion_matrix(y_test, pred, labels=m.classes_)
    display_cm = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=m.classes_)
    display_cm.plot()
    plt.show()
            
    
model_predict(LGBMClassifier)
```

<p>To understand the classification report, we first need to understand this:</p>
<p>&nbsp;</p>
<p><strong>TN / True Negative:&nbsp;</strong>when a case was negative and predicted negative</p>
<p><strong>TP / True Positive:&nbsp;</strong>when a case was positive and predicted positive</p>
<p><strong>FN / False Negative:&nbsp;</strong>when a case was positive but predicted negativ</p>
<p><strong>FP / False Positive:&nbsp;</strong>when a case was negative but predicted positive</p>
<p>&nbsp;</p>
<p><strong>Precision &ndash;&nbsp;</strong>What percent of your predictions were correct?</p>
<p class="wp-block-preformatted"><em><strong>Precision = TP/(TP + FP)</strong></em></p>
<p class="wp-block-preformatted">&nbsp;</p>
<p><strong>Recall &ndash;&nbsp;</strong>What percent of the positive cases did you catch?&nbsp;</p>
<p><em><strong>Recall = TP/(TP+FN)</strong></em></p>
<p>&nbsp;</p>
<p><strong>F1 score&nbsp;&ndash;&nbsp;</strong>What percent of positive predictions were correct? The F1 score is a weighted harmonic mean of precision and recall such that the best score is 1.0 and the worst is 0.0.&nbsp;</p>
<p><em><strong>F1 Score = 2*(Recall * Precision) / (Recall + Precision)</strong></em></p>
